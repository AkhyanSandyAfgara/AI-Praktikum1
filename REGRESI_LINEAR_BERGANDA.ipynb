```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

1. DATA

```python
X = np.array([
    [1, 60],
    [2, 65],
    [3, 70],
    [4, 80],
    [5, 85]
])
y = np.array([50, 55, 65, 75, 85])
```

2. MODEL

```python
model = LinearRegression()
model.fit(X, y)
```

3. PREDIKSI

```python
y_pred = model.predict(X)
```

 4. GRAFIK (asli vs prediksi)

```python
plt.figure()
plt.scatter(range(len(y)), y, label="Asli")
plt.scatter(range(len(y_pred)), y_pred, marker='x', label="Prediksi")

plt.title("Regresi Linear Berganda")
plt.xlabel("Data ke-")
plt.ylabel("Nilai")
plt.legend()
```

Nilai dipengaruhi oleh dua variabel (jam belajar dan kehadiran). Prediksi mendekati nilai asli sehingga model mampu menangkap pengaruh kedua variabel tersebut.
