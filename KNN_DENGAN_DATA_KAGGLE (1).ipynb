```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

1. LOAD DATA (kalau pakai sklearn biar aman)

```python
from sklearn.datasets import load_iris
data = load_iris()

X = data.data[:, :2]   # 2 fitur
y = data.target
```

2. SPLIT DATA

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

 3. MODEL KNN

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

4. PREDIKSI

```python
y_pred = model.predict(X_test)
```

5. GRAFIK

```python
plt.figure()
```

Data asli

```python
plt.scatter(X_test[:,0], y_test, label="Asli")
```

Data prediksi

```python
plt.scatter(X_test[:,0], y_pred, marker='x', label="Prediksi")

plt.title("KNN - Perbandingan Asli vs Prediksi")
plt.xlabel("Fitur 1")
plt.ylabel("Kelas")
plt.legend()
```

Grafik menunjukkan perbandingan antara data asli dan hasil prediksi menggunakan metode KNN. Sebagian besar hasil prediksi mendekati nilai asli, sehingga dapat disimpulkan bahwa model KNN memiliki tingkat akurasi yang cukup baik dalam mengklasifikasikan data.
