```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

1. DATA

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])
```

2. MODEL

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)
```

3. PREDIKSI

```python
y_pred = model.predict(X)
```

4. GRAFIK

```python
plt.figure()
plt.scatter(X, y, label="Asli")
plt.scatter(X, y_pred, marker='x', label="Prediksi")

plt.title("KNN - Data Sendiri")
plt.xlabel("Data")
plt.ylabel("Kelas")
plt.legend()
```

KNN mengklasifikasikan berdasarkan tetangga terdekat (K=3). Hasil prediksi sebagian besar sesuai dengan data asli, menunjukkan metode ini efektif untuk data sederhana.
