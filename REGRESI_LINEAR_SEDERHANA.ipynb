```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

1. DATA

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([45, 50, 60, 65, 75])
```

2. MODEL

```python
model = LinearRegression()
model.fit(X, y)
```

3. PREDIKSI

```python
y_pred = model.predict(X)
```

4. GRAFIK

```python
plt.figure()
plt.scatter(X, y, label="Asli")
plt.plot(X, y_pred, linestyle='dashed', label="Prediksi")

plt.title("Regresi Linear Sederhana")
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai")
plt.legend()
```

Grafik menunjukkan hubungan positif antara jam belajar dan nilai. Garis prediksi mendekati titik asli, menandakan model cukup baik dalam memprediksi nilai berdasarkan satu variabel.
