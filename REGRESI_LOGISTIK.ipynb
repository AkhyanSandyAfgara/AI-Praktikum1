```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

1. DATA

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([0, 0, 0, 1, 1])  # 0 = tidak lulus, 1 = lulus
```

2. MODEL

```python
model = LogisticRegression()
model.fit(X, y)
```

3. PREDIKSI

```python
y_pred = model.predict(X)
```

4. GRAFIK

```python
plt.figure()
plt.scatter(X, y, label="Asli")
plt.scatter(X, y_pred, marker='x', label="Prediksi")

plt.title("Regresi Logistik")
plt.xlabel("Jam Belajar")
plt.ylabel("Kelas")
plt.legend()
```

Model memisahkan data menjadi dua kelas (lulus/tidak). Sebagian besar prediksi sesuai dengan data asli, menunjukkan model cukup baik untuk klasifikasi sederhana.
